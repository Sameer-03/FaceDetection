```python
face_detector=cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
img2=plt.imread('grp.jpg')
faces=face_detector.detectMultiScale(img2,1.06,5)
```

```python
faces
```

```python
cropped_img=[]
for f in faces:
    x,y,w,h=f
    green=(0,255,0)
    cv2.rectangle(img2,(x,y),(x+w,y+h),green,5)
    cropped_img.append(img2[y:y+h,x:x+w])

plt.figure(figsize=(8,8))
plt.imshow(img2)
```

```python
n_lst=[]
for cf in cropped_img:
    r_img=cv2.resize(cf,(100,100))
    n_lst.append(r_img)
```

```python
x=np.hstack(n_lst)
plt.imshow(x)
```
